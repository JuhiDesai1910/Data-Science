```python
import seaborn as sns
```

```python
sns.countplot(x='quality', data=red_train)
```

```python
# add new features: rating.
# quality 1-4 = bed, 5-6 = average, 7-10 = good
conditions = [
    (red_train['quality']>=7),
    (red_train['quality']<=4)
]
rating = ['good', 'bad']
red_train['rating']= np.select(conditions,rating,default='average')
red_train.rating.value_counts()
```

```python
red_train.groupby('rating').mean()
```

```python
correlation = red_train.corr()
plt.figure(figsize=(14,8))
sns.heatmap(correlation, annot=True, linewidths=0, vmin=-1, cmap='RdBu_r')
```

```python
correlation['quality'].sort_values(ascending=False)
```

```python
bx = sns.boxplot(x='quality',y='alcohol',data=red_train)
bx.set(xlabel='Wine Quality', ylabel='Alcohol Percent',title='Alcohol percent in different wine quality types')
```

```python
bx = sns.boxplot(x='rating',y='sulphates',data=red_train)
bx.set(xlabel='Wine Quality', ylabel='Sulphates',title='Sulphates in different wine quality types')
```

```python
bx = sns.violinplot(x='rating',y='citric acid',data=red_train)
bx.set(xlabel='Wine Rating', ylabel='Citric Acid',title='Citric Acid in different wine quality types')
```

```python
bx = sns.swarmplot(x='rating',y='sulphates',data=red_train)
bx.set(xlabel='Wine Quality', ylabel='Sulphates',title='Sulphates in different wine quality types')
```

```python
sns.lmplot(x='alcohol',y='residual sugar',col='rating',data=red_train)
```
